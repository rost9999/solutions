# Task 3: Bayesian Probability Update

Determine the probability of "H" after each of 9 observed flips using Bayes' theorem.

- 4 biased coins with P(H) = [0.12, 0.27, 0.21, 0.96]
- Observed flips: [H, H, H, T, H, T, H, H, H]
- Initial P(H) = 0.39, after first H: P(H) = 0.67

```python
# ймовірності випадання H для кожної монети
coins = [0.12, 0.27, 0.21, 0.96]
flips = ['H', 'H', 'H', 'T', 'H', 'T', 'H', 'H', 'H']

# початкові ймовірності що обрали кожну монету (рівні)
trust = [0.25, 0.25, 0.25, 0.25]
results = []

for flip in flips:
    # оновлюємо ймовірності по Баєсу
    for i in range(4):
        likelihood = coins[i] if flip == 'H' else 1 - coins[i]
        trust[i] *= likelihood

    # нормалізація
    total = sum(trust)
    trust = [t / total for t in trust]

    # P(H) для наступного кидка
    pH = sum(trust[i] * coins[i] for i in range(4))
    results.append(round(pH, 2))

print('Probabilities:', results)
```
